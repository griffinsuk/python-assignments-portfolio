```python
import sympy as sp
#define the symbols
x, y = sp.symbols('x y')
#1.Define the function f (x, y) = ex sin(y)+y3. Compute the first partial derivatives ∂f∂x and ∂f ∂y.
#define the function
f = sp.exp(x) * sp.sin(y) + y**3
df_dy = sp.diff(f, y)
df_dx = sp.diff(f, x)

print (df_dy)
print (df_dx)

#2. gradient vector
x, y = sp.symbols('x y')
g = x**2 * y + x * y**2
grad_g = [sp.diff(g, x), sp.diff(g, y)]
print("Gradient vector =", grad_g)

grad_at_1_1 = [expr.subs({x: 1, y: 1}) for expr in grad_g]
print("Gradient at (1,1):", grad_at_1_1)

#c)
h = sp.ln(x**2 + y**2)
h_xx = sp.diff(h, x, x)  # ∂²h/∂x²
h_yy = sp.diff(h, y, y)  # ∂²h/∂y²
h_xy = sp.diff(h, x, y)  # ∂²h/∂x∂y
print("∂²h/∂x²:", h_xx)
print("∂²h/∂y²:", h_yy)
print("∂²h/∂x∂y:", h_xy)

#The function exhibits a saddle-like shape.

#d)
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
k = sp.sin(x**2 + y**2)

k_func = sp.lambdify((x, y), k, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = k_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = \sin(x^2 + y^2)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
